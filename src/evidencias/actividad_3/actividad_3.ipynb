## Ejercicio 1
Crear un DataFrame frutas con las cantidades de Granadilla y Tomates.


```python
import pandas as pd
import os

class Ejercicios:
    def __init__(self):
        datos = [(i, None) for i in range(1, 13)]
        self.df = pd.DataFrame(data=datos, columns=["#ejercicio", "valor"])
        self.ruta_raiz = os.path.abspath(os.getcwd())
        self.ruta_Actividad_3 = self.ruta_raiz
        os.makedirs(self.ruta_Actividad_3, exist_ok=True)

    def ejercicio1(self):
        frutas = pd.DataFrame({"Granadilla": [20], "Tomates": [50]})
        ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_1.csv")
        frutas.to_csv(ruta, index=False)
        self.df.loc[0, "valor"] = ruta
        print("Ejercicio 1:")
        print(frutas)
        print("********************************************************************")
```

## Ejercicio 6
Figura 42. Utiliza el m√©todo `.info()` para averiguar cu√°ntas entradas hay en el DataFrame.

### Explicaci√≥n
El m√©todo `.info()` de Pandas muestra informaci√≥n general del DataFrame:  
- N√∫mero de entradas (filas).  
- N√∫mero de columnas.  
- Tipos de datos de cada columna.  
- Cantidad de valores no nulos por columna.  

Esto nos permite confirmar la dimensi√≥n del dataset y la calidad de los datos.


```python
def ejercicio6(self):
    import io
    review = pd.read_csv("wine_review.csv")  # Ajusta el nombre/ruta seg√∫n tu archivo
    buffer = io.StringIO()
    review.info(buf=buffer)  # enviamos la salida al buffer
    info_str = buffer.getvalue()
    
    print("Ejercicio 6:")
    print(info_str)  # mostramos la informaci√≥n en pantalla
    
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_6.csv")
    with open(ruta, "w") as f:
        f.write(info_str)  # guardamos la informaci√≥n en un archivo
    self.df.loc[5, "valor"] = ruta
    print("********************************************************************")

Ejercicios.ejercicio6 = ejercicio6
```

## Ejercicio 7
Figura 43. ¬øCu√°l es el precio promedio?

### Explicaci√≥n
Usamos el m√©todo `.mean()` de Pandas sobre la columna `price` para calcular el precio promedio de los vinos en el dataset.  
Este valor nos da una idea general del costo medio de las botellas registradas.


```python
def ejercicio7(self):
    review = pd.read_csv("wine_review.csv")  # Ajusta el nombre/ruta seg√∫n tu archivo
    promedio = review["price"].mean()
    print("Ejercicio 7:")
    print("Precio promedio:", promedio)
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_7.csv")
    # Guardamos el resultado en un archivo CSV
    pd.DataFrame({"PrecioPromedio": [promedio]}).to_csv(ruta, index=False)
    self.df.loc[6, "valor"] = ruta
    print("********************************************************************")

Ejercicios.ejercicio7 = ejercicio7
```

## Ejercicio 8
Figura 44. Obtener el precio m√°s alto de los vinos en el dataset.

### Explicaci√≥n
Usamos el m√©todo `.max()` de Pandas sobre la columna `price` para encontrar el valor m√°ximo registrado en el dataset.  
Este resultado nos muestra cu√°l fue el precio m√°s alto pagado por una botella de vino.


```python
def ejercicio8(self):
    review = pd.read_csv("wine_review.csv")  # Ajusta el nombre/ruta seg√∫n tu archivo
    maximo = review["price"].max()
    print("Ejercicio 8:")
    print("Precio m√°s alto:", maximo)
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_8.csv")
    # Guardamos el resultado en un archivo CSV
    pd.DataFrame({"PrecioMaximo": [maximo]}).to_csv(ruta, index=False)
    self.df.loc[7, "valor"] = ruta
    print("********************************************************************")

Ejercicios.ejercicio8 = ejercicio8
```

Figura 42. Crea un DataFrame con todos los vinos de California

## Ejercicio 10
Encontrar el vino con el precio m√°s alto en el dataset.

### Explicaci√≥n
Usamos `idxmax()` para localizar el √≠ndice del precio m√°ximo y luego `.loc[]` para mostrar toda la fila correspondiente.


```python
def ejercicio10(self):
    review = pd.read_csv("wine_review.csv")  # Ajusta el nombre/ruta seg√∫n tu archivo
    idx = review["price"].idxmax()  # √≠ndice del vino m√°s caro
    vino_caro = review.loc[idx]     # toda la informaci√≥n de ese vino
    print("Ejercicio 10:")
    print(vino_caro)
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_10.csv")
    vino_caro.to_frame().T.to_csv(ruta, index=False)  # guardamos como DataFrame
    self.df.loc[9, "valor"] = ruta
    print("********************************************************************")

Ejercicios.ejercicio10 = ejercicio10
```

## Ejercicio 11
Contar las variedades de vinos de California en el dataset.

### Explicaci√≥n
Aqu√≠ filtramos los registros por la provincia `California` y usamos `.value_counts()` para obtener la frecuencia de cada variedad.


```python
def ejercicio11(self):
    review = pd.read_csv("wine_review.csv")
    california = review[review["province"] == "California"]
    comunes = california["variety"].value_counts()
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_11.csv")
    comunes.to_csv(ruta)
    self.df.loc[10, "valor"] = ruta
    print("Ejercicio 11:")
    print(comunes.head())
    print("********************************************************************")

Ejercicios.ejercicio11 = ejercicio11
```

## Ejercicio 12
Mostrar las 10 variedades de vinos m√°s comunes en California.

### Explicaci√≥n
Aqu√≠ filtramos los registros por la provincia `California` y usamos `.value_counts().head(10)` para obtener las 10 variedades m√°s frecuentes.


```python
def ejercicio12(self):
    review = pd.read_csv("wine_review.csv")  # Ajusta el nombre/ruta seg√∫n tu archivo
    california = review[review["province"] == "California"]
    top10 = california["variety"].value_counts().head(10)
    print("Ejercicio 12:")
    print(top10)
    ruta = os.path.join(self.ruta_Actividad_3, "Ejercicio_12.csv")
    top10.to_csv(ruta)
    self.df.loc[11, "valor"] = ruta
    print("********************************************************************")

Ejercicios.ejercicio12 = ejercicio12
```
