### Ejercicio 5 
Se utiliza la librería Matplotlib para graficar una función matemática simple. Se genera un conjunto de valores `x` y se calcula `y = x²`. Luego se representa la relación en una gráfica de línea.

### Ejercicio 6 
Se define una clase `Rectángulo` con atributos `ancho` y `alto`. La clase incluye métodos para calcular el área y el perímetro. Se crea una instancia de la clase y se muestran los resultados.

```python
# Ejercicio 6: Clase Rectángulo con métodos para área y perímetro 
class Rectangulo: 
    def __init__(self, ancho, alto): 
       self.ancho = ancho 
       self.alto = alto 
       
    def area(self): 
        return self.ancho * self.alto 

    def perimetro(self): 
        return 2 * (self.ancho + self.alto) 
        
# Crear una instancia y mostrar resultados
mi_rectangulo = Rectangulo(5, 3) 
print("Área:", mi_rectangulo.area()) 
print("Perímetro:", mi_rectangulo.perimetro())
```

### Ejercicio 12
Se genera un array `x` con valores entre -2π y 2π.  
Se calcula `y = sin(x) + ruido gaussiano` y se grafica en un gráfico de dispersión.  
También se grafica la curva original `y = sin(x)` para comparar ambas funciones.


```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 12: Gráfico de dispersión con función seno y ruido gaussiano
x = np.linspace(-2*np.pi, 2*np.pi, 200)
y_ruido = np.sin(x) + np.random.normal(0, 0.2, size=x.shape)  # ruido gaussiano
y_sin = np.sin(x)

plt.figure(figsize=(8, 5))
plt.scatter(x, y_ruido, color='red', alpha=0.6, label='sin(x) + ruido')
plt.plot(x, y_sin, color='blue', linewidth=2, label='sin(x)')
plt.title('Función seno con ruido gaussiano')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

### Ejercicio 13
Se utiliza `np.meshgrid` para crear una cuadrícula de valores en los ejes X e Y.  
Luego se aplica la función `z = cos(x) + sin(y)` y se genera un gráfico de contorno con Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 13: Gráfico de contorno con np.meshgrid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.cos(X) + np.sin(Y)

plt.figure(figsize=(8, 6))
plt.contour(X, Y, Z, cmap='viridis')
plt.title('Gráfico de contorno: z = cos(x) + sin(y)')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='Valor de z')
plt.show()
```

### Ejercicio 14
Se generan 1000 puntos aleatorios en dos arrays `x` y `y`.  
Se calcula la densidad de los puntos utilizando un histograma bidimensional y se asigna un color a cada punto según esa densidad.  
Finalmente, se crea un gráfico de dispersión coloreado por densidad.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde

# Ejercicio 14: Gráfico de dispersión con densidad de puntos
x = np.random.rand(1000)
y = np.random.rand(1000)

# Calcular densidad de puntos
xy = np.vstack([x, y])
densidad = gaussian_kde(xy)(xy)

plt.figure(figsize=(7, 6))
plt.scatter(x, y, c=densidad, cmap='viridis', s=30, alpha=0.7)
plt.title('Gráfico de dispersión con densidad de puntos')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(label='Densidad')
plt.show()
```

### Ejercicio 15
Se genera una cuadrícula de valores `x` e `y` con `np.meshgrid`.  
Luego se aplica la función `z = sin(x) + ruido gaussiano` y se genera un gráfico de contorno lleno (`contourf`) para visualizar la distribución de valores.


```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 15: Gráfico de contorno lleno con función seno + ruido
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(x, y)

# Función con ruido gaussiano
Z = np.sin(X) + np.random.normal(0, 0.2, size=X.shape)

plt.figure(figsize=(8, 6))
contorno = plt.contourf(X, Y, Z, cmap='plasma')
plt.title('Gráfico de contorno lleno: sin(x) + ruido')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(contorno, label='Valor de z')
plt.show()
```

Se añaden etiquetas para los ejes, un título y leyendas con formato LaTeX para mejorar la presentación.

### Ejercicio 20
Se definen varias categorías y dos conjuntos de valores.  
Luego se crea un gráfico de barras apiladas para visualizar la composición de cada categoría.


```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 20: Gráfico de barras apiladas
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [3, 5, 2, 6, 4]
valores2 = [2, 3, 4, 2, 5]

plt.figure(figsize=(8, 5))
plt.bar(categorias, valores1, color='lightblue', edgecolor='black', label='Grupo 1')
plt.bar(categorias, valores2, bottom=valores1, color='salmon', edgecolor='black', label='Grupo 2')

plt.title('Gráfico de barras apiladas')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.legend()
plt.grid(axis='y')
plt.show()
```

### Ejercicio 21
Se generan tres series de datos con funciones matemáticas (`sin`, `cos` y `sin+cos`).  
Luego se grafican en un mismo gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

# Ejercicio 21: Gráfico de líneas con múltiples series
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) + np.cos(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$\sin(x)$', color='blue')
plt.plot(x, y2, label=r'$\cos(x)$', color='red')
plt.plot(x, y3, label=r'$\sin(x) + \cos(x)$', color='green')

plt.title('Gráfico de líneas con múltiples series')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
